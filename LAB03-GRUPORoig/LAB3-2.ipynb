# 1. Realiza todo el preprocesamiento necesario para poder entrenar el clasificador y muestra las distintas tablas de distribución.

### Preprocesamiento de los datos

```python
data.info()#no hay missing values
```

```python
data = pd.get_dummies(data) #de variables categoricas a numericas en distintas columnas
data
```

```python
data.columns#Mostramos todas las columnas 
```

```python
print(data['zona_pecho_?'])#Se han creado columnas sin informacion relevante
```

```python
#Mostramos todas las rows de esta columna para demostrar que es irrelevante
pd.set_option('display.max_rows', data.shape[0]+1)
print(data['capsulas_?'])
```

```python
data=data.drop(["capsulas_?",'zona_pecho_?'],axis=1)#Eliminamos las columnas irrelevantes
```

```python
#dado que lo que se han creado dos columnas para radioterapia, una radioterapia_yes y otra radioterapia_no
# y dado que una es la contraria de la otra optaremos por mantener radioterapia_yes
data=data.drop(["radioterapia_no"],axis=1)
```

```python
data #Mostramos los datos tras el preprocesamiento
```

### Tablas distribucion

```python
#for que recorre todas las columnas del dataframe de los datos
#Muestra las tablas de frecuencias con la columna radioterapia_yes con el resto de columnas
#Utilizamos la funcion display para mostrar varios dataframes
#Para mostrar las tablas de distribucion hacemos uso de la funcion de pandas.crosstab
for x in data.columns:
    disp.display(pd.crosstab(index=data[x], columns=data['radioterapia_yes']))
```

```python
#Tablas de distribucion normalizadas por columnas, es decir, divide el count values que podemos ver en la
#anterior tabla entre la suma de los count values de cada columna.
for x in data.columns:
    disp.display(pd.crosstab(index=data[x], columns=data['radioterapia_yes'], normalize='columns'))
```

```python
#Eliminamos 5 rows para el ultimo apartado
test = data.tail(5)
data.drop(data.tail(5).index,inplace=True)
```

# 2. Crea un clasificador e indica su error. Úsalo para saber a qué clase corresponden al menos 5 pacientes que no hayas usado para entrenar el modelo

He decidido utilizar bayes como clasificador dado que en uno de los apartados se nos pide sacar las tablas de distribucion y por ello he entendido que este es el clasificador que se nos pedia.

## Creacion del clasificador e indica su error

```python
#Creamos los datos con de entrenamiento y el test
x=data.drop(["radioterapia_yes"],axis=1)
y=data["radioterapia_yes"]
#80% para el training 20%para el test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x
```

```python
#Creacion del modelo
bayes=GaussianNB()
bayes = bayes.fit(x,y)
#Error y precision del modelo
print('Precision: '+ str(bayes.score(x_test,y_test)))
print('Error: '+ str(1-bayes.score(x_test,y_test)))
```

```python
bayes.class_prior_#Probabilidad de las clases
```

```python
bayes.class_count_#Numero de samples por cada label /target
```

```python
bayes.classes_#Labels/target
```

Podemos ver que hay 164 samples para el target 0 mientras que solo hay 37 samples para el target 1, esto quiere decir que los datos estan desbalanceados y esto puede explicar la poca precision del modelo dado que entrena con muchos mas datos del target 0 que del target 1.

## Clasificar 5 pacientes nuevos

```python
test#Utilizamos los datos que hemos quitado antes del dataset.
```

```python
#Test clasificar 5 rows
x_test=test.drop(["radioterapia_yes"],axis=1)
y_test=test["radioterapia_yes"]
```

```python
#Pasamos los datos nuevos por el clasificador y mostramos la matriz de confusio
confusion_matrix(bayes.predict(x_test), y_test)
```

Podemos ver que clasifica incorrectamente todos los casos positivos, sin embargo, clasifica correctamente los casos negativos.
Esto tiene sentido dado que ha entrenado con muchos mas datos con el target 0 que con el target 1.
